```python
import pandas as pd
import numpy as np

data = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
test
```

```python
data = data.drop(columns = ["Name","Cabin"])
test = test.drop(columns = ["Name","Cabin"])
data
```

```python
data=data.replace('TRAPPIST-1e',0)
data=data.replace('55 Cancri e',1)
data=data.replace('PSO J318.5-22',2)
data=data.replace('Europa',0)
data=data.replace('Earth',1)
data=data.replace('Mars',2)
```

```python
test=test.replace('Europa',0)
test=test.replace('Earth',1)
test=test.replace('Mars',2)
test=test.replace('TRAPPIST-1e',0)
test=test.replace('55 Cancri e',1)
test=test.replace('PSO J318.5-22',2)
```

```python
test=test.fillna(test.mean())
```

```python
 #data=data['Age','RoomService','FoodCourt','Spa','VRDeck','ShoppingMall'].fillna(data.mean(),inplace = True)
data['Age']=data["Age"].fillna(data['Age'].mean())
data['RoomService']=data["RoomService"].fillna(data['RoomService'].mean())
data['FoodCourt']=data["FoodCourt"].fillna(data['FoodCourt'].mean())
data['Spa']=data["Spa"].fillna(data['Spa'].mean())
data['VRDeck']=data["VRDeck"].fillna(data['VRDeck'].mean())
data['ShoppingMall']=data["ShoppingMall"].fillna(data['ShoppingMall'].mean())
data['HomePlanet']=data["HomePlanet"].fillna(1)
data['Destination']=data['Destination'].fillna(0)
data['CryoSleep']=data['CryoSleep'].fillna(0)
data['VIP']=data['VIP'].fillna(0)
data['Transported']=data['Transported'].fillna(0)
data
```

```python
test['Age']=test["Age"].fillna(test['Age'].mean())
test['RoomService']=test["RoomService"].fillna(test['RoomService'].mean())
test['FoodCourt']=test["FoodCourt"].fillna(test['FoodCourt'].mean())
test['Spa']=test["Spa"].fillna(test['Spa'].mean())
test['VRDeck']=test["VRDeck"].fillna(test['VRDeck'].mean())
test['ShoppingMall']=test["ShoppingMall"].fillna(test['ShoppingMall'].mean())
test['HomePlanet']=test["HomePlanet"].fillna(1)
test['Destination']=test['Destination'].fillna(0)
test['CryoSleep']=test['CryoSleep'].fillna(0)
test['VIP']=test['VIP'].fillna(0)
```

```python
X = data.drop(columns=['Transported'])
Y=data['Transported']
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)

rf.fit(X,Y);
```

```python
predictions = rf.predict(test)
predictions
```

```python
predictions=np.round(predictions)
```

```python
from google.colab import files


df = pd.DataFrame(data={'PassengerId': test['PassengerId'], 'Transported':  predictions })
df.to_csv('dataframe.csv')

files.download('dataframe.csv')
```
